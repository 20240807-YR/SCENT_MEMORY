EDA-3. Sensory Language Clustering 비전문 감성 언어는 구조를 가지는가

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.decomposition import PCA

df = pd.read_csv("../data_csv/fra_perfumes.csv", encoding="utf-8")
texts = df["Description"].fillna("").str.lower()
len(texts)
```

```python
import re

tokens = texts.apply(
    lambda x: re.findall(r"[a-z]{4,}", x)
)

tokens.head()
```

```python
all_words = [w for row in tokens for w in row]
word_freq = Counter(all_words)

vocab = [
    w for w, c in word_freq.items()
    if c >= 50
]
len(vocab)

vocab[:20]
```
