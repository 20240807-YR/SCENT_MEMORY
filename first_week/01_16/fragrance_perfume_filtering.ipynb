```python
import pandas as pd

df = pd.read_csv("../data_csv/reviews_0-250.csv", encoding="utf-8")
df.shape
```

```python
perfume_keywords = [
    "perfume", "parfum", "eau de", "edp", "edt",
    "cologne", "fragrance spray"
]

exclude_keywords = [
    "lip", "mask", "cream", "lotion", "cleanser",
    "serum", "toner", "balm", "sleeping"
]
```

```python
name_lower = df["product_name"].fillna("").str.lower()

is_perfume_name = name_lower.apply(
    lambda x: any(k in x for k in perfume_keywords)
)

not_skincare = ~name_lower.apply(
    lambda x: any(k in x for k in exclude_keywords)
)

df_name_filtered = df[is_perfume_name & not_skincare]
df_name_filtered.shape
```

```python
text_lower = df_name_filtered["review_text"].fillna("").str.lower()

context_words = [
    "sillage", "projection", "dry down", "longevity",
    "top notes", "middle notes", "base notes",
    "spray", "wear", "lasting"
]

is_perfume_context = text_lower.apply(
    lambda x: any(w in x for w in context_words)
)

df_frag = df_name_filtered[is_perfume_context]
df_frag.shape
```
