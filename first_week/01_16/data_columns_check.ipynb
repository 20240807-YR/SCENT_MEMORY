```python
import pandas as pd

paths = {
    "final_perfume_data": "../data_csv/final_perfume_data.csv",
    "fra_perfumes": "../data_csv/fra_perfumes.csv",
    "parfumo_datos": "../data_csv/parfumo_datos.csv"
}

dfs = {}

encodings = ["utf-8", "latin1", "cp1252"]

for name, path in paths.items():
    print(f"\n=== {name} ===")
    for enc in encodings:
        try:
            df = pd.read_csv(path, encoding=enc)
            dfs[name] = df
            print(f"✔ loaded with encoding: {enc}")
            print("shape:", df.shape)
            print("columns:")
            for c in df.columns:
                print(" -", c)
            display(df.head(3))
            break
        except UnicodeDecodeError:
            print(f"✖ failed with encoding: {enc}")
```

	•	fra_perfumes → 분포, 평점 구조, 메타 통계
	•	final_perfume_data → 텍스트/임베딩/의미 분석
	•	parfumo_datos → 노트 체계 예시, 도메인 지식 보조
